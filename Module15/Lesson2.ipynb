```python
import numpy as np

def step_function(x):
    return np.where(x >= 0, 1, 0)

class Perceptron:
    def __init__(self, input_size, learning_rate=0.01, epochs=1000):
        self.W = np.zeros(input_size + 1)
        self.learning_rate = learning_rate
        self.epochs = epochs

    def predict(self, x):
        return step_function(np.dot(self.W, x))

    def train(self, x, y):
        for _ in range(self.epochs):
            for xi, target in zip(x, y):
                xi = np.insert(xi, 0, 1)
                prediction = self.predict(xi)
                self.W += self.learning_rate * (target - prediction) * xi

# Данные для задачи XOR Меняем целевые метки (y) на те, что соответствуют задаче XOR
x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])  # Операция XOR

# Попробуем обучить на 10,000, 20,000 и 50,000 эпохах
for epochs in [10000, 20000, 50000]:
    print(f"\nОбучение однослойного перцептрона на {epochs} эпохах:")
    perceptron = Perceptron(input_size=2, epochs=epochs)
    perceptron.train(x, y)
    # Тестирование
    for xi in X:
      xi_with_bias = np.insert(xi, 0, 1)
      print(f"{xi} -> {perceptron.predict(xi_with_bias)}")
```
